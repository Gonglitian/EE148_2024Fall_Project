### DeepLabv3 with ResNet50 backbone and ResNet101 backbone

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res_50_acc = np.load(r'result\resnet50_pretrain\global_acc_list.npy')
res_50_iou = np.load(r'result\resnet50_pretrain\iou_list.npy')
res_50_train_loss = np.load(r'result\resnet50_pretrain\train_loss_list.npy')
res_50_val_loss = np.load(r'result\resnet50_pretrain\val_loss_list.npy')

res_101_acc = np.load(r'result\resnet101_pretrain\global_acc_list.npy')
res_101_iou = np.load(r'result\resnet101_pretrain\iou_list.npy')
res_101_train_loss = np.load(r'result\resnet101_pretrain\train_loss_list.npy')
res_101_val_loss = np.load(r'result\resnet101_pretrain\val_loss_list.npy')
```

```python
#plot accuracy with grid
plt.figure()
plt.plot(res_50_acc, label='ResNet50')
plt.plot(res_101_acc, label='ResNet101')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid()
plt.legend()
plt.show()
```

```python
# plot iou with grid
plt.figure()
plt.plot(res_50_iou, label='ResNet50')
plt.plot(res_101_iou, label='ResNet101')
plt.title('IoU')
plt.xlabel('Epoch')
plt.ylabel('IoU')
plt.grid()
plt.legend()
plt.show()
```

```python
# plot train loss and val_loss with grid
plt.figure()
plt.plot(res_50_train_loss, label='ResNet50_train_loss')
plt.plot(res_50_val_loss, label='ResNet50_val_loss')
plt.plot(res_101_train_loss, label='ResNet101_train_loss')
plt.plot(res_101_val_loss, label='ResNet101_val_loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.legend()
```
